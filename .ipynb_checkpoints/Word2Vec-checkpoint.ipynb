```python
import pandas as pd
from preprocess import preprocess_recipes

df = pd.read_json("../data/train.json").set_index("id")
df['cuisine'] = df.cuisine.astype("category")
df = preprocess_recipes(df)
```

## Word2Vec
Word2Vec is a model used to produce word embeddings. We treat a recipe as a sentence, and an individual ingredient as a word. From Word2Vec, we get vector embeddings for ingredients.

```python
import gensim

def gensim_list(df):
    result = []
    for index,row in df.iterrows():
        result.append(gensim.utils.simple_preprocess (row['ingredients']))
    return result

documents = gensim_list(df)
model = gensim.models.Word2Vec(
        sentences = documents,#the documents
        size=60, #The size of the dense vector to represent each token or ingredient
        window=100, # The maximum distance between the target word and its neighboring word
        min_count=2, #Minimium frequency count of words
        workers=10)
model.train(documents, total_examples=len(documents), epochs=20)

model.save(open('models/word2vec.model', 'wb'))

word_vectors = model.wv
```

Lets see how usefull the embeddings are. Gensim offers a few simple vector operations that can suggest if the learned embeddings make sense to us.

```python
# Word that does not match in a sequence

ingredient_sets =[
    "peach apple banana peanuts",
    "peach pork beef chicken",
    "cheese milk beef butter"
]
for ingredient_set in ingredient_sets:
    print(word_vectors.doesnt_match(ingredient_set.split(' ')))
```

```python
# Similarity between two words
word_touples = [('chicken', 'beef'),('blueberri', 'raspberri'),('cola_soft_drink','sugar'), ('vegan_chees', 'chees'),
                ('rice', 'black_bean'), ('rice','chicken'), ('avocado', 'oliv_oil'), ('avocado','banana')]
similarities = list(zip(word_touples, [word_vectors.similarity(i[0],i[1]) for i in word_touples]))
for i in similarities:
    print(i)
```

```python
# most_similar performs the calculation
# Olive oil - olives + sesame = sesame oil ?
word_vectors.most_similar(positive=['oliv_oil', 'sesam'], negative=['oliv'])
```

## Recipe embeddings
Because we have an embedding for each ingredient, a representation of a recipe will simply be an average of the ingredient embeddings.

```python
import numpy as np
word_vectors = {key: np.array(model.wv.word_vec(key)) for key in model.wv.vocab.keys()}
def encode_recipes(X):
    X_new = []
    for sentence in list(X):
        l = sentence.split(' ')
        vec = [word_vectors.get(i,np.zeros(60)) for i in l]
        X_new.append(sum(vec)/ len(vec))
    return np.array(X_new)
X_encoded = encode_recipes(df['ingredients'])
#word_vectors
```

## Visualising the learned embeddings
We use PCA and TSNE to reduce the learned embeddings to two dimensions, and then plot them if we see any clusters.

```python
%%time

from sklearn.pipeline import Pipeline
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
cutdown = 5000

pipeline = Pipeline([('pca', PCA(n_components=20)),
                     ('tsne', TSNE(n_components=2))])
X_for_plot = pipeline.fit_transform(X_encoded)
```

```python
import numpy as np
#labels = np.reshape(np.array(df['cuisine'][:cutdown]),(cutdown,1))
labels = np.reshape(np.array(df['cuisine']),(39774, 1))
data = np.hstack((X_for_plot, labels))
df_to_plot = pd.DataFrame(data, columns=['x','y','cuisine'])
df_to_plot.head()
```

```python
df_to_plot.to_csv('plot_values.csv')
```

```python
from matplotlib import pyplot as plt
import pandas as pd
plotdata = df_to_plot
fig = plt.figure(figsize=[10,10])
ax = fig.add_subplot(1, 1, 1)
 
for name,group in plotdata.groupby(['cuisine']):
    ax.plot(group['x'], group['y'], marker='o', linestyle='', ms=3, label=name, alpha=0.25)
plt.title('Cuisines scatter plot')
plt.legend(loc=4)
plt.show
plt.savefig('cuisines.png')
```

We can see clusters ! yaay

## Neural network
Lets see how well a classifier performs with the learned embeddings

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_encoded, pd.get_dummies(df['cuisine']), random_state=3628800, test_size=0.2)
model = Sequential()
model.add(Dense(activation="relu", input_dim=60, units=100, kernel_initializer="uniform"))
model.add(Dropout(0.5, noise_shape=None, seed=None))

model.add(Dense(activation="softmax", units=20, kernel_initializer="uniform"))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy', 'top_k_categorical_accuracy'])
model.fit(X_train, y_train, batch_size = 32, epochs = 15, verbose=0)

score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
